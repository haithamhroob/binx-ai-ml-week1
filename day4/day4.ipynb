# Task 2: Handle Missing Values

The dataset is inspected for missing values to identify incomplete records. Appropriate cleaning is applied before continuing with the analysis.

# Task 3: Filter the Data

The dataset is filtered based on selected conditions to focus on a specific subset of vehicles. This makes the analysis more targeted and meaningful.

```python
import pandas as pd
from pandas import DataFrame , Series
import numpy as np
df = pd.read_csv("exp1_14drivers_14cars_dailyRoutes.csv" , low_memory=False)
col = ["MARK" , "MODEL" , "CAR_YEAR" , "SPEED" ,"ENGINE_COOLANT_TEMP" , "ENGINE_RPM" , "AIR_INTAKE_TEMP"]
cars = df[col].copy()
pest = cars[(cars["SPEED"] > cars["SPEED"].mean()) & (cars["ENGINE_RPM"] > 1100) & (cars["ENGINE_COOLANT_TEMP"] < 95)]
pest = pest.sort_values(by="SPEED", ascending=False)
print(pest)
```

The filtered data shows vehicles running above the average speed while keeping the coolant temperature below 95.

# Task 4: Group and Summarize the Data

The filtered data is grouped by vehicle manufacturer to calculate summary statistics. This helps compare the performance of different car brands.

```python
import pandas as pd
from pandas import DataFrame , Series
import numpy as np
df = pd.read_csv("exp1_14drivers_14cars_dailyRoutes.csv" , low_memory=False)
col = ["MARK" , "MODEL" , "CAR_YEAR" , "SPEED" ,"ENGINE_COOLANT_TEMP" , "ENGINE_RPM" , "AIR_INTAKE_TEMP"]
cars = df[col].copy()
pest = cars[(cars["SPEED"] > cars["SPEED"].mean()) & (cars["ENGINE_RPM"] > 1100) & (cars["ENGINE_COOLANT_TEMP"] < 95)]
pest.sort_values(by=["SPEED" , "ENGINE_RPM"] , ascending=False)
group = pest.groupby("MARK").agg(average_speed =("SPEED" , "mean") , Max_Speed =("SPEED", "max") , Average_RPM=("ENGINE_RPM" , "mean") , Average_Coolant=("ENGINE_COOLANT_TEMP","mean") , Records=("MARK" , "count"))
summery = group.sort_values("average_speed" , ascending=False)
summery.style.highlight_max(color="lightgreen")
print(summery)
```

The grouped results help compare the average speed, RPM, and coolant temperature between the available vehicle brands.
