```python
import pandas as pd
from pandas import DataFrame , Series
import numpy as np
df = pd.read_csv("/home/haitham/binx-ai-ml-internship/week1/day4/exp1_14drivers_14cars_dailyRoutes.csv")
print(df.shape)
print(df.head())
print(df.describe())
print(df.isnull().sum())

col = ["MARK" , "MODEL" , "CAR_YEAR" , "SPEED" ,"ENGINE_COOLANT_TEMP" , "ENGINE_RPM" , "AIR_INTAKE_TEMP"]
cars = df[col].copy()
print(cars)

pest = cars[(cars["SPEED"] > cars["SPEED"].mean()) & (cars["ENGINE_RPM"] > 1100) & (cars["ENGINE_COOLANT_TEMP"] < 95)]
print(pest)

pest.sort_values(by=["SPEED" , "ENGINE_RPM"] , ascending=False)


group = pest.groupby("MARK").agg(average_speed =("SPEED" , "mean") , Max_Speed =("SPEED", "max") , Average_RPM=("ENGINE_RPM" , "mean") , Average_Coolant=("ENGINE_COOLANT_TEMP","mean") , Cars=("MARK" , "count"))
summery = group.sort_values("average_speed" , ascending=False)
summery.style.highlight_max(color="lightgreen")
print(summery)
```
