# Task 3: Fillter The Data

The dataset is filtered based on selected conditions to focus on a specific subset of vehicles. This makes the analysis more targeted and meaningful.

```python
import pandas as pd
from pandas import DataFrame , Series
import numpy as np
df = pd.read_csv("/home/haitham/binx-ai-ml-internship/week1/day4/exp1_14drivers_14cars_dailyRoutes.csv" , low_memory=False)
col = ["MARK" , "MODEL" , "CAR_YEAR" , "SPEED" ,"ENGINE_COOLANT_TEMP" , "ENGINE_RPM" , "AIR_INTAKE_TEMP"]
cars = df[col].copy()
pest = cars[(cars["SPEED"] > cars["SPEED"].mean()) & (cars["ENGINE_RPM"] > 1100) & (cars["ENGINE_COOLANT_TEMP"] < 95)]
pest.sort_values(by=["SPEED" , "ENGINE_RPM"] , ascending=False)
print(pest)
```

# Task 4: Group and Summarize the Data

The filtered data is grouped by vehicle manufacturer to calculate summry statistics. This helps compare the performance of different car brands.

```python
import pandas as pd
from pandas import DataFrame , Series
import numpy as np
df = pd.read_csv("/home/haitham/binx-ai-ml-internship/week1/day4/exp1_14drivers_14cars_dailyRoutes.csv" , low_memory=False)
col = ["MARK" , "MODEL" , "CAR_YEAR" , "SPEED" ,"ENGINE_COOLANT_TEMP" , "ENGINE_RPM" , "AIR_INTAKE_TEMP"]
cars = df[col].copy()
pest = cars[(cars["SPEED"] > cars["SPEED"].mean()) & (cars["ENGINE_RPM"] > 1100) & (cars["ENGINE_COOLANT_TEMP"] < 95)]
pest.sort_values(by=["SPEED" , "ENGINE_RPM"] , ascending=False)
group = pest.groupby("MARK").agg(average_speed =("SPEED" , "mean") , Max_Speed =("SPEED", "max") , Average_RPM=("ENGINE_RPM" , "mean") , Average_Coolant=("ENGINE_COOLANT_TEMP","mean") , Cars=("MARK" , "count"))
summery = group.sort_values("average_speed" , ascending=False)
summery.style.highlight_max(color="lightgreen")
print(summery)
```
