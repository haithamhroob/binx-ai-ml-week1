# Data Visualization with Matplotlib

In this notebook, I practiced the basic plotting techniques using Matplotlib and the Iris dataset from Seabron.
The notebook covers the four core plot types and how to organize multiple plots using subplots.

```python
import matplotlib.pyplot as plt
import seaborn as sns
flower = sns.load_dataset("iris")
print (flower.head())
```

## Bsic Plot Types

The following examples demonstrate the main visualization techniques using the Iris dataset.

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes[0,0].plot(flower['sepal_length'] , flower['sepal_width'],marker='o',c='tab:brown' ,lw='2', ls='-' ,mfc='orange' , mec='lime', mew='2' , alpha=.5)
axes[0,0].set_title("Line Plot")
axes[0,0].set_xlabel("Speal Length")
axes[0,0].set_ylabel("Speal Width")
axes[0,0].legend(["mark"])

axes[0,1].scatter(flower['petal_length'] , flower["petal_width"] , color='tab:red')
axes[0,1].set_title("Scatter Plot")
axes[0,1].set_xlabel("Petal Length")
axes[0,1].set_ylabel("Petal Width")

species_mean = flower.groupby("species")["sepal_length"].mean()

axes[1 , 0].bar(species_mean.index , species_mean.values , color=["royalblue" , "orange" ,"green"])
axes[1,0].set_title("Average sepal Length")
axes[1,0].set_xlabel("Species")
axes[1,0].set_ylabel("Length")

axes[1 , 1].hist(flower["sepal_length"],bins=15,color="purple")
axes[1,1].set_title("Sepal Length Distibution")
axes[1,1].set_xlabel("Sepal Length")
axes[1,1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

## Summary

This notebook demonstrates the four basic Matplotlib plot types:
- Line Plot
- Scatter Plot
- Bar Plot
- Histogram

The plots were organized using subplots to compare multiple visualizations in a single figure.
