## Step 1 : Function for Basic statistics

This function takes a list of numbers and calculates: the mean , minimum , and maximum values. The results are returned as a dictionary ,making them easy to access and use in later analysis.

```python
def Calculation(numbers):
    total = sum(numbers)
    average = total / len(numbers) if numbers else 0
    maximum = max(numbers) if numbers else None
    minimum = min(numbers) if numbers else None
    return {
        "average": average,
        "maximum": maximum,
        "minimum": minimum
    }
    
nums = [34 , 255 , 11 , 32 , -23 , 311 , 23]
result = Calculation(nums)
print(result)
    
```

## Step 2 : List Comprehensions 

This example demonstrates how to use a list comprehension to filter even numbers from a list. List comprehnsions provide a concise and readable way to create new lists in Python.

```python
numbers = [1 , 2 , 3 , 4]
evens = [x for x in numbers if x % 2 == 0]
for x in evens:
    print(x)
```

## Step 3 : Data record class

This example defines a simple class that represents a data record. The class contains two attributes to store personal information: the name and age, and one method to display the stored information. This demonstrates the basic concepts of OOP in Python.

```python
class personal_info:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        return f"Name: {self.name}, Age: {self.age}"

n = "haitham"
a = 20
data = personal_info(n, a)
print(data.display_info())
```
