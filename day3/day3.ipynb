# Final Mini Project

## Vehicle Sensor Analysis
This mini project analysez vehicle OBD-II sensor readings using NumPy

### Features
- Load sensor readings from a CSV file
- Compute statistical information
- Detect driving characteristics
- Generate a formatted analysis report
- Save the report automatically as a text file

## Why this project?
I intentionally selected automotive sensor data because it directly relates to my academic specialization in IOT&Embedded Systems and supports the direction of my graduation project.

Rather than treating the exercise as a standalone programming task, I wanted to practice newly learned cocepts in a context that reflects my long-term enginnering interests.

## Output

- console report
- 'vehicle_report.txt'


```python
import numpy as np

np.set_printoptions(precision=2, suppress=True)

sensors_data = np.genfromtxt('../../OBDii_data.csv', delimiter=',', skip_header=1)
sensors_data =np.delete(sensors_data, 0, axis=1)
print(sensors_data)
header = ["Engine_RPM", "Vehicle_Speed_KMH", "Throttle_Pos_Pct", "Coolant_Temp_C", "Engine_Load_Pct"]

max_speed = np.max(sensors_data[:,1])
mean_speed = np.mean(sensors_data[:,1])
min_speed = np.min(sensors_data[:,1])
speed_mps = sensors_data[: , 1] / 3.6

max_rpm = np.max(sensors_data[:,0])
min_rpm = np.min(sensors_data[:,0])
mean_rpm = np.mean(sensors_data[:,0])

max_coolant_temp = np.max(sensors_data[:,3])
min_coolant_temp = np.min(sensors_data[:,3])
mean_coolant_temp = np.mean(sensors_data[:,3])

max_load = np.max(sensors_data[:,4])
min_load = np.min(sensors_data[:,4])
mean_load = np.mean(sensors_data[:,4])


how_many_speed_over_mean = np.sum(sensors_data[:,1] > mean_speed)
how_many_rpm_over_2000 = np.sum(sensors_data[:,0] > 2000)
highest_speed_record = sensors_data[np.argmax(sensors_data[:,1])]

if mean_speed < 30:
    driving_condition = "City Driving"
elif mean_speed < 60:
    driving_condition = "Suburban Driving"
else:
    driving_condition = "Highway Driving"

report = f"""
============================================================================================
                                         Vehicle Sensor Data Report
============================================================================================
Driving Condition: {driving_condition}


Max Speed: {max_speed:.2f} km/h
Mean Speed: {mean_speed:.2f} km/h
Min Speed: {min_speed:.2f} km/h


Max RPM: {max_rpm:.2f}
Mean RPM: {mean_rpm:.2f}
Min RPM: {min_rpm:.2f}


Max Coolant Temp: {max_coolant_temp:.2f} °C
Mean Coolant Temp: {mean_coolant_temp:.2f} °C
Min Coolant Temp: {min_coolant_temp:.2f} °C


Max Engine Load: {max_load:.2f} %
Mean Engine Load: {mean_load:.2f} %
Min Engine Load: {min_load:.2f} %


Number of Speed Records Over Mean Speed: {how_many_speed_over_mean}

Number of RPM Records Over 2000: {how_many_rpm_over_2000}

Highest Speed Record: {header[1]}: {highest_speed_record[1]:.2f} km/h, {header[0]}: {highest_speed_record[0]:.2f} RPM, {header[2]}: {highest_speed_record[2]:.2f} %, {header[3]}: {highest_speed_record[3]:.2f} °C, {header[4]}: {highest_speed_record[4]:.2f} %    

speed in m/s: {speed_mps}

"""
print(report)
with open("vehicle_sensor_report.txt", "w") as report_file:
    report_file.write(report)
```
